## Задание 3

### Изучите взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров, используя сложенную или многорядовую столбчатую диаграмму. Процессоры от Samsung не изучайте

### 3.1 Постройте график в абсолютных величинах

```python
data = pd.crosstab(index = df['Company'], columns = df['Cpu'])
#data = data.drop(columns = ['Samsung']).reset_index()
data
```

```python
import numpy as np
plt.figure(figsize = (8,6))
n_ticks = np.arange(len(data['AMD A12-Series 9720P 3.6GHz']))
offset = 0.2

plt.bar(n_ticks - offset, data['AMD A12-Series 9720P 3.6GHz'], width = 0.4)
plt.bar(n_ticks + offset, data['Intel Pentium Dual Core 4405U 2.1GHz'], width = 0.4)

plt.title('Ноутбуки')
plt.xlabel('Компания - производитель ноутбуков')
plt.ylabel('Количество ноутбуков')
plt.legend(['AMD A12-Series 9720P 3.6GHz', 'Intel Pentium Dual Core 4405U 2.1GHz'])
plt.xticks(n_ticks, data['Company'], rotation = 50) 
```

### 3.2 Постройте график в относительных величинах

```python
data = pd.crosstab(index = df['Company'], columns = df['Cpu'], normalize = 'index')
```

```python
import numpy as np
plt.figure(figsize = (8,6))

plt.bar(data['Company'], data['AMD A12-Series 9720P 3.6GHz'])
plt.bar(data['Company'], data['Intel Pentium Dual Core 4405U 2.1GHz'], bottom = data['AMD A12-Series 9720P 3.6GHz'])

plt.title('Ноутбуки')
plt.xlabel('Компания - производитель ноутбуков')
plt.ylabel('Доля ноутбуков')
plt.legend(['AMD A12-Series 9720P 3.6GHz', 'Intel Pentium Dual Core 4405U 2.1GHz'])
plt.xticks(rotation = 50) 
```

### Постройте матрицу корреляций для таблицы

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(corr_matrix, annot = True, linewidths = 0.5, cmap = 'coolwarm')
```
