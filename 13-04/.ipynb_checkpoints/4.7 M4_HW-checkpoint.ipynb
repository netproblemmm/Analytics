<div class="alert alert-info">
* Правда ли женщины в среднем имеют более высокую зарплату, чем мужчины?
  Образец:

```python
compare_1_1 = data.loc[data['sex'] == 0, ['salary']]
compare_1_2 = data.loc[data['sex'] == 1, ['salary']]
```

```python
bins = 50
plt.figure(figsize=(6, 6))
plt.hist(compare_1_1['salary'], bins=bins, alpha=0.9, edgecolor="black", label="Womens")
plt.hist(compare_1_2['salary'], bins=bins, alpha=0.7, label="Mens")
plt.legend(loc="upper right")
plt.show()
```

```python
t_value, p_value = mannwhitneyu(compare_1_1['salary'], compare_1_2['salary'])
 
print('Статистики=%.3f, p=%.3f' % (t_value, p_value))
 
alpha = 0.05
if p_value > alpha: # Интерпретируем
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
```

<div class="alert alert-info">
* Правда ли, что инженеры (Engineer) зарабатывают в среднем больше, чем люди из сферы развлечений (Entertainment)?
  ОБразец:

```python
compare_2_1 = data.loc[data['profession'] == 'Engineer', ['salary']]
compare_2_2 = data.loc[data['profession'] == 'Entertainment', ['salary']]
```

```python
bins = 50
plt.figure(figsize=(6, 6))
plt.hist(compare_2_1['salary'], bins=bins, alpha=0.9, edgecolor="black", label="Engineer")
plt.hist(compare_2_2['salary'], bins=bins, alpha=0.7, label="Entertainment")
plt.legend(loc="upper right")
plt.show()
```

```python
t_value, p_value = stats.ttest_ind(compare_1_1['salary'], compare_1_2['salary'])
 
print('Статистики=%.3f, p=%.3f' % (t_value, p_value))
 
alpha = 0.05
if p_value > alpha: # Интерпретируем
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
```

<div class="alert alert-info">

<div class="alert alert-info">
Правда ли, что женщины с зарплатой выше медианной в среднем более счастливые, чем мужчины с зарплатой, которая выше средней?

```python
def calculate_median(data):  
    sorted_data = sorted(data)  
    n = len(sorted_data)  
    mid = n // 2  
    if n % 2 == 0:  
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2  
    else:  
        return sorted_data[mid]

compare_7_1 = data.loc[data['sex'] == 1, ['salary']]
print(calculate_median(compare_7_1))
```
